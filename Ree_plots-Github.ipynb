```python
# Basic packages used in the plots. 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style("darkgrid")
colors = ["tab:blue","tab:red","tab:orange","tab:green","tab:purple", "tab:pink",
          "tab:olive", "tab:gray", "tab:brown"] #colors used by the plots
colors_dashed = ["b", "r", "darkorange","g", ]
size = 28 #font size for some plots
```

```python
df=pd.read_csv("./data_reeM.csv", sep = ";")
```

```python
#Selection and lop to create the list of elements in the plot
list(df.columns)
elements = list(df.columns) 
elements.remove("Material")
elements.remove("Sample")
elements.remove("Site")


ele = []
for e in elements:
        ele.append(e)
         
elements=ele
        
```

```python
# First plot and area of test. 

fig, ax = plt.subplots(1, figsize=(15,10))

pd.plotting.parallel_coordinates(df, cols = elements, class_column = "Material", ax = ax, color = colors,
                             linewidth=2.5, markersize=1)
plt.yscale("log")
plt.ylim(10**0.1,10**4.5)
plt.ylabel('REE$_{sample}$/REE$_{Chondrite}$', fontsize=20)
#plt.title('REE LA-ICP-MS',fontsize=16)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(loc='upper right', fontsize=18)
```

```python
#Full data plot
fig, ax = plt.subplots(1, figsize=(15,10))


pd.plotting.parallel_coordinates(df, cols = elements, class_column = "Material", ax = ax, color = colors,
                             linewidth=2.5, markersize=8)
plt.yscale("log")
plt.ylim(10**0.1,10**4.5)
plt.ylabel('REE$_{sample}$/REE$_{Chondrite}$', fontsize=14)
plt.title('REE LA-ICP-MS',fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(loc='upper right', fontsize=12.5)
```

```python
# new df to have mean, min and maximun values 
result = df.groupby('Material').agg({e: ['mean', 'min', 'max'] for e in elements})
result = result.stack(level = 1).reset_index().rename(columns = {"level_1": "type"})
result
```

```python
mask1 = (df["Material"] == "Ore") | (df["Material"] == "Sandstone") | (df["Material"] == "Bloom") | (df["Material"] == "Oven Slag") | (df["Material"] == "Plate Slag") | (df["Material"] == "Flow Slag") | (df["Material"] ==  "Iron-SI")
mask2_1 = (result["Material"] == "Oven Slag") | (result["Material"] == "Plate Slag") |(result["Material"] == "Flow Slag") |(result["Material"] == "Smithing Slag")
mask3 = (result["type"] != "mean") 


fig, ax = plt.subplots(1, figsize=(15,10))

pd.plotting.parallel_coordinates(result[mask3], cols = elements, class_column = "Material", ax = ax, 
                                 color=colors, linewidth=2.8, markersize=8)
plt.yscale("log")
plt.ylim(10**0.1,10**4.5)
plt.ylabel('REE$_{sample}$/REE$_{Chondrite}$', fontsize=15)
plt.title('All Materials (Max & Min)',fontsize=20)
ax.legend(loc='upper right', fontsize=12.5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
fig.savefig("./plot2/full_mm.png", dpi=200)
```

```python
maskmean = (result["type"] == "mean") 

fig, ax = plt.subplots(1, figsize=(15,10))

pd.plotting.parallel_coordinates(result[maskmean], cols = elements, class_column = "Material", ax = ax, 
                                 color= colors, linewidth=3, markersize=10)
plt.yscale("log")
plt.ylim(10**0.25,10**4.5)
plt.ylabel('REE$_{sample}$/REE$_{Chondrite}$', fontsize=15)
plt.title('REE LA-ICP-MS Results (Mean)',fontsize=20)
ax.legend(loc='upper right', fontsize=12.5)
plt.xticks(fontsize=14)
plt.yticks(fontsize=12)
plt.tight_layout()
fig.savefig("./plot2/full_mm.png", dpi=200)
```

```python
#New data frame from the df and results 
df1 = df.loc[(df["Material"]=="Ore")]
df2 = result.loc[(result["Material"].isin(["Iron-SI","Iron-SI-P", "Bloom", "Flow Slag", "Oven Slag", "Plate Slag", "Smithing Slag"])) & (result["type"] == "mean") ]
```

```python
# append process to create a new data frame conving the two previus 
df_lines = df1.append(df2,ignore_index=True).fillna("")
df_lines
```

```python
# final plot using the new data frame convining samples by materail and samples by site (Ores)
mask1 = (df_lines["Material"] == "Ore")
mask2 = (df_lines["Material"] != "Ore" ) 
#ores = {"Eisenberg": "red", "Imbasch": "yellow", "Hettenleidelheim": "cyan","Tiefenthal": "purple", "Stupfwald": "green" }

#elements_sig = list(df_lines[elements].max()[df[elements].max() > 0.25].index)
fig, ax = plt.subplots(1, figsize=(15,10))


pd.plotting.parallel_coordinates(df_lines[mask2], cols = elements, class_column = "Material", ax = ax, color=colors,
                             linewidth=3, markersize=1)

pd.plotting.parallel_coordinates(df_lines[mask1], cols = elements, class_column = "Site", ax = ax, colormap = "brg",
                             linewidth=3,linestyle = "dashed", markersize=1)
plt.yscale("log")
plt.ylim(10**0.25,10**4.5)
plt.ylabel('REE$_{sample}$/REE$_{Chondrite}$', fontsize=20)
plt.title('Rare Earth Elements  Results',fontsize=25)
ax.legend(loc='upper right', fontsize=12.5)
plt.xticks(fontsize=14)
plt.yticks(fontsize=12)
plt.tight_layout()
fig.savefig("./plot2/full-ree-material-ore.png", dpi=200)
```
